<p> <strong> Introduction:</strong> Malnutrition is a condition of deficiencies, excesses, or imbalances in a person's intake of energy and/or nutrients (WHO, 2024). It includes undernutrition, inadequate vitamins or minerals, overweight, obesity, and resulting diet-related noncommunicable diseases (WHO, 2024). Malnutrition affects millions of people worldwide and can result in severe outcomes including a weakened immune system, increased vulnerability to infections, stunted growth and developmental delays (BAPEN, 2023). It can also cause detrimental long-term effects to an individuals physical and mental health. Children, infants and adolescents are disproportionately impacted by this condition, as well as those living in under-resourced low income communities. </p>

<p> This data analysis on global malnutrition focuses on the rates of undernutrition, speficially severe wasting, wasting, stunting, overweight, and underweight data. Wasting is an indicative factor for low weight to height, stunting indicates low height for age, reflecting delayed growth and undernutrition, and underweight is a composite measure indicating low weight for age, as well as wasting and stunting included (WHO 2024). </p>

<p> This data set outlines the global rates of these malnutrition indicators among children between the ages of 0-59 months (under 5 years old), from 1983-2019. It also measure indicative factors of income classification (low, lower middle, upper middle or high). The goal of this analysis is to study trends in malnutrition rates globally, and identify key measures among different countries. By analyzing these trends, public health professionals will have greater insight into areas suffering the most from this condition, and can work towards creating effective change to reduce the rates of malnutrition. </p>

```python
# Compare the means of Severe Wasting between the top 10 and bottom 10 countries
# Get the top 10 countries with the highest rates of Severe Wasting
top_10_severe_wasting = df.nlargest(10, 'Severe Wasting')['Severe Wasting']
# Get the bottom 10 countries with the lowest rates of Severe Wasting
bottom_10_severe_wasting = df.nsmallest(10, 'Severe Wasting')['Severe Wasting']
# Perform a t-test to compare the means of the two groups
t_stat, p_value = stats.ttest_ind(top_10_severe_wasting, bottom_10_severe_wasting)
# Print the t-statistic and p-value
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')
```

<div style="background-color:rgb(245, 187, 136); color: black; padding: 15px; border-radius: 8px; font-family: 'Times New Roman', Times, serif;">

  <p>The t-statistic and p-value indicate whether there is a significant difference in the means of the two groups. The mean value of Severe Wasting in the top 10 countries is significantly higher than that in the bottom 10 countries, indicating that this distinction is not due to chance. This highlights the potential differences in healthcare and income classification levels between countries, that may contribute to severe wasting (extremely low weight to height rates). The countries with lower rates of severe wasting also have higher income classification levels, suggesting that socioeconomic status effects malnutrition rates.</p>
